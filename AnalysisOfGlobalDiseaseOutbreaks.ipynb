## EDA Analysis of Reported cases of Disease Outbreaks in the World
With this dataset, I will be exploring and analyzing the effect and reported cases of different diseases in the World. I will also be identifying trends and comparing my results of my analysis. 

```python
## Step 1: load all neccessary libraries for EDA 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
import os 
```

```python
## Step 2: Upload and import dataset 
df = pd.read_csv('C:/Users/Valerie/OneDrive/CAPSTONE PROJECT/PandemicData_Topic7/1- the-number-of-cases-of-infectious-diseases.csv')
```

```python
## Step 3: View shape, size, # of cols, etc. 
df.head()
```

```python
## Displaying info of created dataframe
df.info()
```

```python
# Detecting missing values and calculating mean of each col. 
df.isna() .mean()
```

```python
## Step 4: sorting cols. in ascending order
df = df.sort_values(by=["Year", "Entity"], axis=0, ascending=True)
```

```python
df.head()
```

```python
## Step 5: Rename Columns 
df.rename(columns={
    "Indicator:Number of cases of yaws reported": "yaws",
    "Total polio cases": "polio",
    "Reported cases of guinea worm disease in humans": "guinea worm",
    "Number of new cases of rabies, in both sexes aged all ages": "rabies",
    "Number of new cases of malaria, in both sexes aged all ages": "malaria",
    "Number of new cases of hiv/aids, in both sexes aged all ages": "hiv/aids",
    "Number of new cases of tuberculosis, in both sexes aged all ages": "tuberculosis",
    "Number of reported smallpox cases": "smallpox",
    "Reported cholera cases": "cholera"
}, inplace=True)
```

```python
df['Year'] = df['Year'].astype(int) .astype(str)
```

```python
## Step 6: Dropping Null values in Cols. 
df.dropna(subset=['yaws'], inplace=True)
df.dropna(subset=['Total (estimated) polio cases'], inplace=True)
df.dropna(subset=['guinea worm'], inplace=True)
df.dropna(subset=['rabies'], inplace=True)
df.dropna(subset=['malaria'], inplace=True)
df.dropna(subset=['hiv/aids'], inplace=True)
df.dropna(subset=['tuberculosis'], inplace=True)
df.dropna(subset=['smallpox'], inplace=True)
df.dropna(subset=['cholera'], inplace=True)
```

```python
# Displaying rows of table 
df.head()
```
